```python
import pandas as pd
import numpy as np
from datetime import datetime

#Load Import File
data_fp = str(input("Enter data path: /Users/bkanchipurambaburavi/Downloads/BOMTEMP.xlsx"))
data_file = pd.ExcelFile(data_fp)
hbom = pd.read_excel(data_file, sheet_name='Sheet1')
print("Ok")
```

```python
hbom_rename = {'tpn':'Tesla PN', 'mfr':'Manufacturer', 'mpn':'MPN', 'component type':'Type', 'qty':'Qty', 'harness list':'Harness PN','description':'Description'}
hbom = hbom.rename(columns=hbom_rename)
column_name = list(hbom.columns)

print(column_name)
```

```python
abc = []
len_file = len(hbom.iloc[:,1])
for i in range (len_file):
    tsla_h = hbom.iloc[i, 6].split(";")
    abc.append(tsla_h)
print(abc)
```

```python
final_array = []

final_array.append(column_name)

for i in range(len(abc)):
    z = len(abc[i])
    a = hbom.iloc[i,0]
    c = hbom.iloc[i,1]
    d = hbom.iloc[i,2]
    e = hbom.iloc[i,3]
    f = hbom.iloc[i,4]
    g = hbom.iloc[i,5]
    for j in range(z):
        b = abc[i][j]
        final_array.append([a,c,d,e,f,g,b])

print(final_array)
print(len(final_array))
```

```python
#Write dataframes to Excel file and save to local drive
fp = str(input("Please enter output file name:\n"))
#savefp = input("Indicate filepath to save file in:\n")

final = pd.DataFrame(final_array)

#generating separate component level forecast sheet for each suppliers
with pd.ExcelWriter(fp + '.xlsx') as writer:
    final.to_excel(writer, index=False, header=False, sheet_name='Sheet1')
    print('Saved')    
```
